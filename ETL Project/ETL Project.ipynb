```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
csv_file = "economic_freedom.csv"
freedom_data_df = pd.read_csv(csv_file)
freedom_data_df.head()
```

```python
new_freedom_data = freedom_data_df[['FREEDOM ID','Country name','index year', 'overall score','business freedom','trade freedom','investment freedom','financial freedom']].copy()
new_freedom_data.head()
```

```python
v2_freedom_data= new_freedom_data[pd.notnull(new_freedom_data['overall score'])]
v2_freedom_data.head()
```

```python
csv2_file = "world_trade_banking.csv"
trading_data_df = pd.read_csv(csv2_file)
trading_data_df.head()
```

```python
trading_data = trading_data_df[['TRADE ID','LOCATION','TIME','Value']].copy()
trading_data.head()
```

```python
csv3_file="countrycode.csv"
countrycode_df = pd.read_csv(csv3_file)
countrycode_df.head()
```

```python
new_trading_data = pd.merge(trading_data,countrycode_df, on="LOCATION")
new_trading_data.head()
```
